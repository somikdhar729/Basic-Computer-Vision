# Decoding hidden messages from Images
### Detective Sherlock left a message for his assistant Dr. Watson while tracking his arch-enemy Professor Moriarty. Could you help Dr. Watson decode this message? 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
image_dir = 'E:/Computer Vision Practice/Images/'
```

```python
img = cv2.imread(image_dir + 'for_watson.png')
```

```python
cv2.imshow('',img)

# Wait for the user to press a key
cv2.waitKey(0)
 
# Close all windows
cv2.destroyAllWindows()
```

```python
np.unique(img,return_counts=True)
```

```python
rgb = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
cv2.imshow('',rgb)

# Wait for the user to press a key
cv2.waitKey(0)
 
# Close all windows
cv2.destroyAllWindows()
```

```python
plt.imshow(rgb);
plt.title("Original Image");
```

To plot using matplotlib as OpenCV uses BGR notation

```python
np.unique(rgb,return_counts=True)
```

As we can see that majority of the values are either 0 or 255.

Possible reasons why the message may be hidden:
* The color of the texts is same as the background color<br>

Possible solution:
* Change the background color<br>
    * Remove the background

```python
img3 = rgb
```

```python
img3[img3 == 0] = 255 
```

```python
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
cv2.imshow('',img3)

# Wait for the user to press a key
cv2.waitKey(0)
 
# Close all windows
cv2.destroyAllWindows()
```

```python
plt.imshow(img3);
plt.title("Hidden Message");
```
